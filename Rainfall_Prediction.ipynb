**NOW START DATA PREPROCESSING**

```python
# drop multicolinear features
## Drop maxtemp , temparature , mintemp AND keep=  dewpoint because of this highly impect on rainfall feature
# also drop day feature because we are not working on time series

df = df.drop(['maxtemp' , 'temparature' , 'mintemp' , 'day'] , axis=1)
```

```python
df.head()
```

```python
# are target classes are imbalanced if result are not good then we make it balance

# now divide our data for training and testing

X = df.drop('rainfall' , axis = 1)
y = df['rainfall']
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X ,y , test_size=0.20 , random_state=42)
```

```python
# now scale the feature for better performance
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# now it's time for TRAING THE MODEL
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train , y_train)
```

```python
y_pred_knn = knn.predict(X_test)
y_pred_train_knn = knn.predict(X_train)
```

## AFTER TRAINING THE MODEL
### EVALUTE MODEL


```python
from sklearn.metrics import accuracy_score , f1_score , classification_report , confusion_matrix , precision_score , recall_score
```

```python
print("accuracy training" , accuracy_score(y_test , y_pred_knn))
print("accuracy testing" , accuracy_score(y_train , y_pred_train_knn) )
```

```python
print("precision training")
print(precision_score(y_test , y_pred_knn))
print("precision testing" , precision_score(y_train , y_pred_train_knn) )

# recall
print('recal testing' ,recall_score(y_test , y_pred_knn))
print("recall training" , recall_score(y_train , y_pred_train_knn) )
```

```python
f1 = f1_score

print("f1 test ")
print(f1(y_test , y_pred_knn))
print("f1 train")

print(f1(y_train , y_pred_train_knn))
```

```python
print("test")
print(confusion_matrix(y_test , y_pred_knn))
```

```python
print(classification_report(y_test , y_pred_knn))
```

```python
#AFTER SCALING CHECK KNN PERFORMANCE

knn_scale = KNeighborsClassifier(n_neighbors=5)

knn_scale.fit(X_train_scaled , y_train)
```

```python
y_pred_scale_knn = knn_scale.predict(X_test_scaled)
y_pred_train_scale_knn = knn_scale.predict(X_train_scaled)
```

```python
print("accuracy training" , accuracy_score(y_test , y_pred_scale_knn))
print("accuracy testing" , accuracy_score(y_train , y_pred_train_scale_knn) )
```
